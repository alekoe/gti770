```python
from matplotlib.colors import ListedColormap
train_path='Simpsons-Train-Valid/'
valid_path='Simpsons-Train-Valid/'

cm = plt.cm.RdBu
colors = ['red','green','blue','purple', 'magenta', 'yellow', 'black', 'cyan' ]
# choices = {'bart' : 0, 'home': 1, 'lisa': 2, 'magg' : 3, 'marg' : 4, 'fami' : 5, 'othe': 6, 'scho' : 7}
cm_bright = ListedColormap(colors)

orange, white, label = np.loadtxt("simpsons_train.csv", delimiter=",", unpack = True)

# load feature file
data = np.loadtxt("simpsons_train.csv", delimiter=",")

graph = plt.scatter(orange, white, c=label, cmap=cm_bright, edgecolors='k', label="Features")
plt.xlabel("# pixels orange")
plt.ylabel("# pixels white")
plt.show()
```

```python
# Just for two classes
# red = Bart
leg = ['red = Bart','green = Homer']

cm = plt.cm.RdBu
colors = ['red','green']
cm_bright = ListedColormap(colors)
data2 = data[np.where( (data[:,2] <= 1.0) )]
graph = plt.scatter(data2[:,0], data2[:,1], c=data2[:,2], cmap=cm_bright, edgecolors='k', label=leg)
plt.xlabel("# pixels orange")
plt.ylabel("# pixels white")
plt.legend()
plt.show()
```
